## Analyzing the correlation between the features in the dataset

```python
# Import the libraries
import pandas as pd
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import metrics
```

```python
# Loading the CSV file from the file location dataset
df = pd.read_excel('Data.xlsx')

# Printing the overview of the raw dataset
df.head()

```

```python
df.info()
```

```python
# Checking the .corr function to see the correlations between variables. closest to one
print(df.corr())
```

```python
# Heat map of all variables

plt.figure(figsize=(16, 8), dpi=300)  # Set the resolution to 300 dpi for high quality
sns.heatmap(df.corr(), cmap="Blues", annot=True)
plt.savefig("heatmap.pdf", format="pdf", bbox_inches="tight")  # Save the figure as a PDF
plt.show()
```

```python
# Interpretation


# A correlation matrix provides a comprehensive view of the relationships between variables in a dataset
# by displaying the correlation coefficients between every pair of variables.
# Each cell in the matrix represents the correlation coefficient between two variables, ranging from -1 to 1.
# A correlation of 1 indicates a perfect positive correlation, meaning that as one variable increases,
# the other variable also increases proportionally. 
# A correlation of -1 indicates a perfect negative correlation, meaning that as one variable increases,
# the other variable decreases proportionally. 
# A correlation of 0 indicates no linear relationship between the variables.

# Interpreting a correlation matrix involves examining the strength and direction of the correlations.
# Strong positive correlations (closer to 1) suggest that the variables move in the same direction, 
# while strong negative correlations (closer to -1) suggest that the variables move in opposite directions. 
# Weak correlations (closer to 0) suggest a lack of linear relationship between the variables.
```
